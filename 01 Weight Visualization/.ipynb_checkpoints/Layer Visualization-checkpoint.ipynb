# Let's start with a standard CNN for MNIST

```python
from keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train[...,None]/x_train.max() # add axis and divide by max to scale [0,1]
x_test = x_test[...,None]/x_test.max()

N_CAT = y_train.max() +1
```

```python
DROPOUT = .1

model_layers = [
    keras.layers.InputLayer(input_shape=x_train.shape[1:]),
    
    keras.layers.Conv2D(20, 3, padding="same", activation="elu"),
    keras.layers.MaxPool2D(padding="same"),
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(DROPOUT),
    
    keras.layers.Conv2D(30, 3, padding="same", activation="elu"),
    keras.layers.MaxPool2D(padding="same"),
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(DROPOUT),
    
    keras.layers.Conv2D(40, 3, padding="same", activation="elu"),
    keras.layers.MaxPool2D(padding="same"),
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(DROPOUT),
    
    keras.layers.GlobalAvgPool2D(),
    
    keras.layers.Dense(N_CAT, activation="softmax")
]
model = keras.Sequential(model_layers)
model.summary()
```

```python
callbacks = [
    keras.callbacks.ReduceLROnPlateau(
        factor=0.5, patience=1, cooldown=1, verbose=1),
    keras.callbacks.EarlyStopping(patience=5, verbose=1)
]
model.compile("nadam", "sparse_categorical_crossentropy", metrics=["acc"])

model.fit(
    x_train,
    y_train,
    validation_data=(x_test, y_test),
    epochs=100,
    callbacks=callbacks)
```
